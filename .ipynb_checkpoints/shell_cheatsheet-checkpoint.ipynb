```python
\enter #换行
mkdir -p #多层
mv name1 name2#重命名
mv name1 name2 dir3 #mv多个文件，最后一个为目录
head -n 10 name#查看前10行
less name1#查看文件，/可搜索字符串
touch name1#创建文件
which -a python #寻找所有的指令目录
locate name#搜索文件名
gzip #压缩
#等号两边不能接空格
#双引号内的特殊字符如 $ 等， 可以保有原本的特性
#单引号内的特殊字符则仅为一般字符
#version=$（ uname -r）
# $（指令）
#“PATH="$PATH":/home/bin”或“PATH=${PATH}:/home/bin”
# 因为子程序仅会继承父程序的环境变量， 子程序不会继承父程序的自订变量啦！ 父程序的自订变量是无法在子程序内使用的。 但是通过 export 将变量变成环境变量后， 就能够在子程序下面应用了
#grep区分大小写
 history -w #强制立刻写入bash_history
!66 #执行history第 66 条指令
source #读入环境配置文件的指令
```
