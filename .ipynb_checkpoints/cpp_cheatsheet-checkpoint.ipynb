```python
#include  <iostream>
#include <fstream>
 //读取txt文件1
ifstream readFileadj("../../../src/slam_graph/data/T_adjacent.txt",ios::in); //以文本模式打开txt备读
if(!readFileadj) { //打开失败
    cout << "error opening T_adjacent.txt file." << endl;
    //return 0;
}
int pushtimes=0;
vector<double> onerow(36,0);
while (!readFileadj.eof())
{
    pushtimes++;
    double temp;
    readFileadj>>temp;
    onerow[pushtimes-1]=temp;
}
```
